```python
from sklearn.datasets import make_classification
```

```python
x,y = make_classification (n_samples=10000,
                           n_features=20,
                           n_informative=14,
                           n_redundant=6,
                           n_classes=2,
                           random_state=42)
```

```python
import tensorflow as tf
import keras
def create_model():
  model=keras.Sequential()
  model.add(keras.layers.Dense(units=512,activation='relu'))
  model.add(keras.layers.Dense(units=10,activation='relu'))
  model.add(keras.layers.Dense(units=1,activation='sigmoid'))

  model.compile(optimizer='adam',loss=keras.losses.binary_crossentropy,metrics=['accuracy'])
  return model
```

```python
n_split = 5
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score

kf=StratifiedKFold(n_splits=n_split, shuffle=True)
```

```python

acc_total = []

for train_index , validation_index in kf.split(x,y):

  x_train = x[train_index]
  y_train = y[train_index]
  x_validation = x[validation_index]
  y_validation = y[validation_index]
  model = create_model()

  model.fit(x_train , y_train ,epochs=64,batch_size=1000,validation_data=(x_validation,y_validation), verbose=2)

  y_pred=model.predict(x_validation)
  y_pred=(y_pred>0.5).astype('int32')

  acc=accuracy_score(y_true=y_validation, y_pred=y_pred)
  acc_total.append(acc)
```

```python
acc_total
import numpy as np
np.mean(acc_total)
```
