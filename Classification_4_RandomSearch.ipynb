```python
from sklearn.datasets import make_classification
x,y = make_classification(n_samples=15000, n_features=20, n_informative=10, n_redundant=5, n_classes=2, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from tensorflow import keras
def model_create(hidden_unit=16, optimizer='adam'):
    model = keras.Sequential()
    model.add(keras.Input(shape=(20,)))
    model.add(keras.layers.Dense(units=hidden_unit, activation='relu'))
    model.add(keras.layers.Dense(units=1, activation='sigmoid'))
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
from scikeras.wrappers import KerasClassifier
model= KerasClassifier(model_create, epochs=10, batch_size=1000, verbose=0)
```

```python
params = {
    'model__hidden_unit':[16,64,256],
    'model__optimizer':['adam','rmsprop']
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
rs = RandomizedSearchCV(model, param_distributions=params, cv=3, n_iter=5)
rs_result= rs.fit(x_train,y_train)
```

```python
print("Best params:", rs_result.best_params_)
print("Best CV accuracy:", rs_result.best_score_)

test_acc = rs_result.score(x_test, y_test)
print("Test accuracy:", test_acc)
```
