```python
import tensorflow as tf
import numpy as np
import keras
from keras import layers
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
#data definition
x=np.arange(-80,81,0.5)
y=x**2+5*x+3

x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model_1=keras.Sequential()
model_1.add(keras.layers.Input(shape=[1]))
model_1.add(keras.layers.Dense(1024,activation='relu'))
model_1.add(keras.layers.Dense(1))
model_1.summary()
```

```python
model_1.compile(optimizer='adam',loss='mean_squared_error')
```

```python
hist_1=model_1.fit(x_train,y_train,batch_size=32,epochs=1000,validation_data=(x_test,y_test))
```

```python
model_2=keras.Sequential()
model_2.add(keras.layers.Input(shape=[1]))
model_2.add(keras.layers.Dense(64,activation='relu'))
model_2.add(keras.layers.Dense(256,activation='relu'))
model_2.add(keras.layers.Dense(256,activation='relu'))
model_2.add(keras.layers.Dense(256,activation='relu'))
model_2.add(keras.layers.Dense(1))

model_2.summary()
```

```python
model_2.compile(optimizer='adam',loss='mean_squared_error')
```

```python
hist_2=model_2.fit(x_train,y_train,batch_size=64,epochs=1000,validation_data=(x_test,y_test))
```

```python
y_pred_1=model_1.predict(x_test)
y_pred_2=model_2.predict(x_test)
```
